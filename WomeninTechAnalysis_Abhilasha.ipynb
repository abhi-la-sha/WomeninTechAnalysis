```python
def verify_missing_fields(female_values, male_values, female_index, male_index):

    fields_in_m = np.setdiff1d(male_index, female_index)
    female_index = list(female_index)
    female_values = list(female_values)
    for i in fields_in_m:
        female_index.append(i)
        female_values.append(0)

    fields_in_f = np.setdiff1d(female_index, male_index)
    male_index = list(male_index)
    male_values = list(male_values)
    for j in fields_in_f:
        male_index.append(j)
        male_values.append(0)

    return female_values, male_values, female_index, male_index
```

```python
def age(fname):

    age1 = fname["q2Age"][(fname["q2Age"] != '#NULL!')].value_counts()
    age = age1.to_numpy()
    age_index = age1.index.to_numpy()

    return age, age_index

age_m, age_index_m = age(file4_male)
age_f, age_index_f = age(file4_female)

age_f, age_m, age_index_f, age_index_m= verify_missing_fields(age_f, age_m, age_index_f, age_index_m)

df_plot = pd.DataFrame({'age_male': age_m,
                   'age_female': age_f}, index=age_index_f)

graph = df_plot.plot.bar(rot=90)
```

```python
def stats(value_m, value_f):

    percent_f = [(value_f[i]/(value_f[i]+val))*100
    for i, val in enumerate(value_m)]
    return percent_f
```

```python
percent_f = stats(age_m, age_f)

df = pd.DataFrame({'percentage of women': percent_f}, index=age_index_f)
graph1 = df.plot.bar(rot=90)
```
